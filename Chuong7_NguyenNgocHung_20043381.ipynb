```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb
print(sb.get_dataset_names())
```

```python
flights = sb.load_dataset("flights")
flights.head()
```

# Thống Kê Số Lượng Hành Khách Mỗi Năm ( Tính theo từng tháng )

```python
pl.figure(figsize=(14, 9))

sb.lineplot(data = flights,
             x="month",
             y="passengers",
             hue = "year",
             palette = "muted"
             )

pl.title("Thống Kê Số Lượng Hành Khách Mỗi Năm")
pl.show()
```

```python
flights_wide = flights.pivot(index="year", columns="month", values="passengers")
flights_wide.head()
```

```python
pl.figure(figsize=(14, 9))
sb.barplot(data=flights_wide)
pl.title("Thống Kê Số Lượng Hành Khách Mỗi Năm")
pl.show()
```

```python
pl.figure(figsize=(14, 9))
sb.scatterplot(data=flights_wide)
pl.title("Thống Kê Số Lượng Hành Khách Mỗi Năm")
pl.show()
```

```python
pl.figure(figsize=(14, 9))
sb.lineplot(data=flights_wide)
pl.title("Thống Kê Số Lượng Hành Khách Mỗi Năm")
pl.show()
```

```python
pl.figure(figsize=(14, 9))
sb.histplot(data=flights_wide)
pl.title("Thống Kê Số Lượng Hành Khách Mỗi Năm")
pl.show()
```
